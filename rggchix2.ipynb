Для генерации списка прокрученных участником колес будем использовать функцию генерации [случайной последовательности](https://www.random.org/sequences/) на сайте random.org. Сгенерируем последовательность чисел от 1 до 1680 в случайном порядке, для каждого участика возьмем первые n колес из полученной последовательности, где n - количество колес, прокрученных участником за весь сезон.


Напишем функцию для генерации колес прокрученных участником за сезон

```python
import requests

# Входные параметры запроса к random.org
params = {
    'min': 1,
    'max': 1680,
    'col': 1,
    'format': 'plain',
    'rnd': 'new'
}

# Функция генерации колес - генерирует список от 1 до 1680 в случайном порядке и возвращает первые n элементов
def generate_wheels(n):
    # Запрос для Sequence Generator
    response = requests.get('https://www.random.org/sequences/', params=params)
    return [int(num) for num in response.text.strip().split('\n')][:n]
```

```python
# Количество колес прокрученных участником за сезон
chix2_total_wheels = {
    "aLySQuE": 255,
    "Charmy_Games": 139,
    "Ditoshey": 236,
    "HellKoty": 159,
    "HiganBanban": 372,
    "HoriKouma": 166,
    "Koordinata": 313,
    "MariTerryCore": 361,
    "MsJudi": 361,
    "nedrugaya": 116,
    "nyamuras": 409,
    "PeaceDaria": 161,
    "pepelnayaa": 145,
    "RUSTY": 101,
    "sad_moustache": 323,
    "Sieliya_Farres": 523,
    "Zhem": 230,
}
```

## Сгенерируем колеса для каждого участника

```python
chix2_wheels = {}
for key in chix2_total_wheels:
    chix2_wheels[key] = generate_wheels(chix2_total_wheels[key])
print(chix2_wheels)
```

## Посчитаем количество выпадений каждого колеса

```python
#количество выпадений каждого пункта колеса
chix2_wheels_count = {}
for key in chix2_wheels:
    for roll in chix2_wheels[key]:
        if roll in chix2_wheels_count.keys():
            chix2_wheels_count[roll] += 1
        else:
            chix2_wheels_count[roll] = 1

#сортируем список
chix2_wheels_count = dict(sorted(chix2_wheels_count.items(), key=lambda item: item[1], reverse=True))
print(chix2_wheels_count)
```
